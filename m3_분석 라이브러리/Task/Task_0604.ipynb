Task6. 0일부터 10일까지의 시간 범위를 12시간 간격으로 생성하고, 각 간격에 대해 누적 시간을 계산하여 데이터프레임을 작성하세요.

```python
# 시간 범위 생성
timedelta_range = pd.timedelta(start=0, end='100', freq='12H')

# 데이터프레임 생성
df = pd.DataFrame({
    'timedelta': timedelta_range,
    'cumulative_time': [timedelta.total_second() / 3600 for timedelta in timedelta_range]
})

print(df)
```

```python
# 시간 범위 생성
timedelta_range = pd.timedelta_range(start='0 days', end='100 days', freq='12H')

# 데이터프레임 생성
df = pd.DataFrame({
    'timedelta': timedelta_range,
    'cumulative_time': [timedelta.total_seconds() / 3600 for timedelta in timedelta_range]
})

print(df)
```

Task8. 주어진 df에서 월별로 데이터의 합계를 계산하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-15', '2022-02-01', '2022-02-15', '2022-03-01'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# 월별 데이터 집계
df['month'] = df['date'].dt.to_period('M')
df
```

```python
# 월별 합계 구하기
monthly_sum = df.groupby('month')['value'].sum().reset_index()
print(monthly_sum)
```

Task9. 주어진 데이터프레임에서 주말(토요일과 일요일) 데이터를 필터링하세요.

Task10. 주어진 df에서 분기별 데이터의 평균을 계산하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01', '2022-05-01'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# 분기 정보를 추출하여 새로운 열 추가
df['quarter'] = df['date'].dt.to_period('Q')

# 분기별로 그룹화하고 평균 계산
quarterly_mean = df.groupby('quarter')['value'].mean().reset_index()

print(quarterly_mean)
```

```python
print(df.groupby(df['date'].dt.quarter)['value'].mean())
```

Task11. 주어진 df에서 작업 일자(월요일부터 금요일)만 남기고 필터링하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-03', '2022-01-04', '2022-01-05', '2022-01-08', '2022-01-09'],
    'value': [1, 2, 3, 4, 5, 6]
})
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# 작업일자
df['day_of_week'] = df['date'].dt.dayofweek
workdays_df = df[df['day_of_week']<5]
print(workdays_df)
```

```python
# 작업 일자(월요일부터 금요일)만 필터링
workdays = df[df['date'].dt.dayofweek.isin([0, 1, 2, 3, 4])]

print(workdays)
```
