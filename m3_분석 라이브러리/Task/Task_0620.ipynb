```python
price_list = []

for tag in soup.find_all(class_='Price_price__GqXqo'):
    if 'Price_delete__EoEjk' in tag['class']:
        # 할인 전 가격일 경우 처리하지 않음
        continue
    else:
        # 정상적인 가격 데이터 추출
        price = tag.text.strip()
        # '캐시' 문자 제거 후 정수로 변환
        price = int(price.replace('캐시', '').replace(',', ''))
        price_list.append(price)

df_prices = pd.DataFrame({'Price': price_list})
```
