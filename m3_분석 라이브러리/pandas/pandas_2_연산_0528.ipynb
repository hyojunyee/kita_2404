```python
import pandas as pd
st1 = pd.Series({'국어':100,'영어':80,'수학':90})
st2 = pd.Series({'수학':80,'국어':90,'영어':80})

print(st1,'\n')
print(st2)
```

```python
# 두 데이터 프레임의 덧셈
pd.DataFrame(st1 + st2)
```

```python
add = st1 + st2
sub = st1 - st2
mul = st1 * st2
div = round((st1 / st2),2)

df = pd.concat([add,sub,mul,div],axis=1)
df
```

```python
df = pd.concat([add,sub,mul,div],axis=1)
df.columns = ['덧셈','뺄셈','곱셈','나눗셈']
# df.index.name = '과목'
df
```

```python
print(add,type(add))
```

```python
# 데이터 프레임에 리스트로 주면 행으로 들어감.
# 여기서는 인덱스에 넣어줘야된다.
result = pd.DataFrame([add,sub,mul,div],index = ['덧셈','뺄셈','곱셈','나눗셈'])

result
```

```python
import numpy as np
import pandas as pd

st1 = pd.Series({'국어':np.nan,'영어':80,'수학':90})
st2 = pd.Series({'수학':80,'국어':90})

add = st1 + st2
sub = st1 - st2
mul = st1 * st2
div = round((st1 / st2),2)

result = pd.DataFrame([add,sub,mul,div],
                      index = ['덧셈','뺄셈','곱셈','나눗셈'])
result
```

```python
st1 = pd.Series({'국어':np.nan,'영어':80,'수학':90})
st2 = pd.Series({'수학':80,'국어':90})

ad = st1.add(st2, fill_value=0)
su = st1.sub(st2, fill_value=0)
mu = st1.mul(st2, fill_value=0)
di = st1.div(st2, fill_value=0)

df = pd.DataFrame([ad,su,mu,di],
                      index = ['덧셈','뺄셈','곱셈','나눗셈'])
df
```

```python
# reshape는 차원의 개념(행/열)
df1 = pd.DataFrame(np.arange(12.).reshape((3,4)), columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape((4,5)), columns=list('abcde'))
print(df1)
print(df2)
```

```python
df2.loc[1,'b'] = np.nan
print(df1)
print(df2)
```

```python
df1+df2
```

```python
# r을 붙여 계산 인자를 뒤집어 계산
print(df1)
print(df1.radd(1))
print(df1.rsub(2))
print(df1.rmul(3))
print(round(df1.rdiv(4),2))
```

```python
arr = np.arange(12.).reshape((3,4))
print(arr)
print(arr[0])
```

```python
# broadcasting
# 각 행마다 지정 값(arr[0])의 값을 빼준다.
arr - arr[0]
```

```python
# DataFrame과 Series간 연산
frame = pd.DataFrame(np.arange(12.).reshape(4,3),columns=list('bde'),
                    index = ['Utah','Ohio','Texas','Oregon'])
frame
```

#### 그룹 연산
* 복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나누어서 관찰할 수 있으며 이런 방식으로 분할 처리하는 것을 그룹 연산이라 함
* 그룹 연산은 데이터를 집계, 변환, 필터링하는데 효율적이며 판다스 groupby() 메소드를 사용함
* 그룹 객체 만들기(분할)
* 그룹 연산 메소드(적용-결합)
