## OpenCV(오픈소스 컴퓨터 비전)
컴퓨터 비전 및 이미지 처리에 널리 사용되는 오픈소스 라이브러리로, 다양한 이미지 및 비디오 관련 작업을 수행할 수 있습니다. OpenCV를 학습할 때 필요한 중요한 개념과 기능은 크게 여러 영역으로 나눌 수 있습니다.

### 1. **OpenCV 설치 및 환경 설정**
   - OpenCV 설치: Python을 위한 `pip install opencv-python` 명령어를 사용하여 설치할 수 있습니다. 추가로 영상과 오디오 처리 관련 라이브러리 설치가 필요할 수 있습니다.
   - 기본적인 개발 환경 설정: Python IDE나 Jupyter Notebook, 또는 OpenCV를 활용할 다른 개발 도구를 설정합니다.

### 2. **이미지 읽기, 쓰기, 표시**
   OpenCV에서 이미지를 다루기 위한 기본적인 작업입니다.
   - `cv2.imread()`: 이미지를 읽어옵니다.
   - `cv2.imshow()`: 이미지를 화면에 표시합니다.
   - `cv2.imwrite()`: 이미지를 저장합니다.
   - 색상 공간(BGR, RGB, 그레이스케일) 변환: 이미지를 다양한 색상 공간으로 변환하는 방법을 이해합니다.

### 3. **이미지 처리**
   이미지 처리 기법은 컴퓨터 비전의 핵심입니다. OpenCV에서 제공하는 다양한 필터와 변환을 학습합니다.
   - **이미지 크기 조정**: `cv2.resize()` 함수로 이미지를 확대/축소하는 방법을 익힙니다.
   - **이미지 자르기 및 영역 설정**: 특정 영역을 선택하고 잘라내는 방법 (`numpy` 배열 슬라이싱 활용).
   - **회전 및 이동**: `cv2.warpAffine()`을 사용하여 이미지를 회전시키거나 이동시키는 방법.
   - **필터 적용**: 블러링, 샤프닝, 에지 검출 같은 기본 필터 적용 방법 (`cv2.blur()`, `cv2.GaussianBlur()`, `cv2.Canny()`).
   - **이미지 연산**: 덧셈, 뺄셈, 비트 연산 등 두 개 이상의 이미지를 합성하는 방법.

### 4. **색상 기반 이미지 처리**
   색상에 기반한 이미지 처리 기법은 물체 추적이나 필터링에 필수적입니다.
   - **색상 공간 변환**: BGR에서 다른 색상 공간(RGB, Grayscale, HSV 등)으로 변환하는 방법 (`cv2.cvtColor()`).
   - **색상 필터링**: 특정 색상 범위를 설정해 이미지를 필터링하거나 추출하는 방법 (`cv2.inRange()`).

### 5. **이미지 변형**
   이미지의 기하학적 변형을 다룹니다.
   - **Affine Transformations (아핀 변환)**: 이미지의 회전, 이동, 크기 조정, 기울기 변환을 학습합니다.
   - **Perspective Transformations (투영 변환)**: 이미지를 기울여 보이도록 변형하거나, 특정 영역만 선택하여 원근감을 반영하는 변환을 학습합니다.

### 6. **에지 검출 및 윤곽선 찾기**
   에지 검출과 윤곽선 찾기는 컴퓨터 비전에서 중요한 요소 추출 방법입니다.
   - **Canny Edge Detection**: 에지를 검출하는 방법 (`cv2.Canny()`).
   - **Contours (윤곽선) 찾기**: 물체의 윤곽선을 찾고 이를 분석하는 방법 (`cv2.findContours()`, `cv2.drawContours()`).

### 7. **영상 처리**
   OpenCV는 실시간 비디오 스트림과 관련된 다양한 기능도 제공합니다.
   - **비디오 읽기 및 쓰기**: `cv2.VideoCapture()`로 카메라나 비디오 파일을 읽고, `cv2.VideoWriter()`로 비디오를 저장합니다.
   - **프레임 처리**: 각 프레임에 대해 이미지 처리 기술을 적용하는 방법을 익힙니다.
   - **FPS 측정**: 실시간 영상에서 초당 프레임 수(FPS)를 측정하고 최적화하는 방법.

### 8. **물체 탐지 및 추적**
   OpenCV는 사전 학습된 알고리즘을 통해 물체를 탐지하고 추적하는 기능을 제공합니다.
   - **Haar Cascades**: 얼굴, 눈, 자동차 등을 감지할 때 사용하는 사전 학습된 XML 분류기 파일과 이를 활용한 물체 탐지 (`cv2.CascadeClassifier()`).
   - **객체 추적**: 사전 정의된 객체 추적 방법을 사용하여 영상에서 물체를 추적하는 방법 (`cv2.Tracker` 클래스).

### 9. **기본 기계 학습 및 심층 학습 모델 활용**
   OpenCV는 기본적인 머신 러닝 및 딥러닝 모델을 사용할 수 있는 기능을 제공합니다.
   - **딥러닝 모델 불러오기**: OpenCV에서 TensorFlow, Caffe, PyTorch 등의 딥러닝 모델을 불러와서 추론에 사용하는 방법 (`cv2.dnn` 모듈).
   - **HOG + SVM**: Histogram of Oriented Gradients(HOG)와 Support Vector Machine(SVM)을 결합하여 물체를 탐지하는 방법.

### 10. **카메라 캘리브레이션 및 3D 재구성**
   - **카메라 캘리브레이션**: 카메라 렌즈 왜곡을 교정하기 위한 기술 (`cv2.calibrateCamera()`).
   - **3D 재구성**: 여러 장의 이미지를 활용해 3D 포인트를 계산하고, 스테레오 비전을 사용한 3D 재구성 방법 (`cv2.stereoCalibrate()`).

### 11. **컴퓨터 비전 프로젝트 통합**
   OpenCV는 다양한 실제 프로젝트에 적용할 수 있습니다. 다음과 같은 프로젝트를 시도해 볼 수 있습니다:
   - **얼굴 인식**: 실시간으로 웹캠을 통해 얼굴을 감지하고, 얼굴 인식 알고리즘을 적용하는 프로젝트.
   - **차선 감지**: 도로 이미지에서 차선을 감지하는 자율 주행 관련 프로젝트.
   - **모션 탐지**: CCTV 영상에서 움직임을 감지하는 방법.

파일에 적접 저장하는 urlretrieve( )함수와 달리 urlopen 함수는 읽어들인 데이터를 임의의 변수에 저장하여 처리할 수 있습니다.
- urlretrieve는 웹 URL로부터 파일을 직접 다운로드하여 로컬 파일 시스템에 저장하는 데 사용됩니다. 이 방식은 파일을 직접 저장할 필요가 있을 때 편리합니다.
- urlopen은 URL을 열어 데이터에 접근하는 파일과 유사한 객체를 반환합니다. 이를 통해 데이터를 실시간으로 읽고 처리할 수 있으며, 반드시 로컬 파일 시스템에 저장할 필요는 없습니다.

```python
# URL에서 이미지를 다운로드하고, OpenCV로 이미지를 읽어서 matplotlib을 사용해 화면에 출력
# urllib.request 모듈은 웹사이트 데이터 접근
import urllib.request as req
import matplotlib.pyplot as plt
import cv2


url = 'https://image.shutterstock.com/z/stock-photo-happy-cheerful-young-woman-wearing-her-red-hair-in-bun-rejoicing-at-positive-news-or-birthday-gift-613759379.jpg'

req.urlretrieve(url, 'lady.png')    # download

# OpenCV로 읽어 들이기
img = cv2.imread('lady.png')

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))    # OpenCV는 기본적으로 BGR 포멧을 사용하기 때문에, Matplotlib과 호환되도록 RGP 포멧으로 변환
# plt.imshow(img)
plt.show()
```

```python
import cv2
import urllib.request
import matplotlib.pyplot as plt
import numpy as np

url = 'https://image.shutterstock.com/z/stock-photo-happy-cheerful-young-woman-wearing-her-red-hair-in-bun-rejoicing-at-positive-news-or-birthday-gift-613759379.jpg'

# 이미지를 메모리로 직접 로드
reqp = urllib.request.urlopen(url)  # 웹 리소스의 내용을 바이트(byte) 형태로 읽기
# bytearray(....) 함수는 바이트열을 가변적인 바이트 배열로 변환
# np.asarray(....) 함수는 주어진 입력(여기서는 bytearray 객체)을 NumPy 배열로 변환
image = np.asarray(bytearray(reqp.read()), dtype="uint8")

# 메모리에서 이미지 디코드
img = cv2.imdecode(image, cv2.IMREAD_COLOR) # 메모리 상의 이미지 데이터를 디코드

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))    # OpenCV는 기본적으로 BGR 포멧을 사용하기 때문에, Matplotlib과 호환되도록 RGP 포멧으로 변환
# plt.imshow(img)
plt.show()
```

```python
# urlopen을 저장 후 시각화
import urllib.request

url = 'https://image.shutterstock.com/z/stock-photo-happy-cheerful-young-woman-wearing-her-red-hair-in-bun-rejoicing-at-positive-news-or-birthday-gift-613759379.jpg'

savename = 'lady.png'

mem=urllib.request.urlopen(url).read()

with open(savename, mode='wb') as f:
    f.write(mem)
    print("저장 되었습니다.")

img = cv2.imread('lady.png')

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

- OpenCV 방식: 이미지를 배열로 읽어들여 BGR에서 RGB로 변환 후 시각화.
- imageio 방식: 이미지를 바로 numpy 배열로 읽어서 시각화.
- OpenCV로 URL 이미지 읽기: URL에서 이미지를 가져와 바이트 배열로 변환 후 디코딩.

```python
# OpenCV 방식
import cv2
import urllib.request
import matplotlib.pyplot as plt
import numpy as np

# 이미지 url
url = 'http://uta.pw/shodou/img/28/214.png'

reqp = urllib.request.urlopen(url)
image = np.asarray(bytearray(reqp.read()), dtype="uint8")

# 메모리에서 이미지 디코드
img = cv2.imdecode(image, cv2.IMREAD_COLOR)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# plt.imshow(img)
plt.show()
```

```python
import numpy as np
img = np.zeros((512,512,3), np.uint8)   # 검은 이미지 생성
```

```python
# (0,0): 선의 시작점 좌표. 여기서는 이미지의 좌상단 모서리를 나타냄
# (511,511): 선의 끝점 좌표. 여기서는 이미지의 우하단 모서리 가까이를 나타냄(512x512 dlalwldptjsms dngkeks ahtjflrk wjdghkrgl (511,511)에 위치)
# (255,0,0): 선의 색상. OpenCV에서 색상은 BGR 형식으로 지정되므로, 이 값은 빨간색을 나타냄
# 3: 선의 두께
img =cv2.line(img,(0,0),(511,511),(255,0,0),3)
plt.imshow(img)
plt.show()
```

```python
# (500,0)의 오른쪽 상단 지점에서 (0,500)의 왼쪽 아래에 있는 점까지 연결 된 두깨가 있는 (5Pixel) 선
black_canvas = np.zeros((500,500,3), dtype='uint8') # 500x500 픽셀의 크기와 3개의 색상 채널(RGB)을 가지는 검정색 이미지를 생성
color = (0,0,255)
img = cv2.line(black_canvas,(500,0),(0,500),color,5) # 시작점이 (500, 0)이고 끝점이 (0, 500)인 파란색 선
plt.imshow(img)
plt.show()
```

```python
img = cv2.rectangle(img,(400,0),(500,128),(255,0,0),5)
plt.imshow(img)
plt.show()
```

```python
# (450,50): 원의 중심점 좌표
# 50: 원의 반지름 원의 크기를 결정
# (0,255,0,0): 원의 색상을 나타내는 BGR 값
# 두깨를 -1로 설정하면 원의 내부가 완전히 채워짐.
# matplotlib을 사용하여 이미지를 표시하면, BRG 형식의 이미지를 RGB 형식으로 해석하기 때문에, 빨간색이 파란색으로, 파란색이 빨간색으로 보이게 됨.
img = cv2.circle(img,(450,50),50,(0,255,255),-1)
plt.imshow(img)
plt.show()
```

```python
img = cv2.circle(img,(50,450),50,(255,0,0,0),5)
plt.imshow(img)
plt.show()
```

`cv2.ellipse(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType[, shift]]])`

|parameter|내용|
|-|-|
|img|이미지 파일|
|center|타원의 중심 좌표(x, y)|
|axes|축의 절반 길이(중심에서 긴 거리, 짧은 거리)|
|angle|타원의 기울기|
|startAngle|타원을 그리는 시작 각도|
|endAngle|타원을 그리는 종료 각도|
|color|색상 (blue, green, red) 0 ~ 255|
|thickness|선 두께 (default 1)|
|lineType|선 종류 (default cv.Line_8)</br>- LINE_8 : 8-connected line</br>- LINE_4 : 4-connecterd line</br>- LINE_AA : antialiased line|
|shift|fractional bit (default 0)|
