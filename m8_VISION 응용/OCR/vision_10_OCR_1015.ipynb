## **Tesseract OCR**
- 오픈 소스 OCR(Optical Character Recognition, 광학 문자 인식) 라이브러리로, 이미지나 PDF 파일에서 텍스트를 추출하는 데 사용됩니다. 이 라이브러리는 문자 인식의 정확성과 강력함으로 인해 다양한 애플리케이션에서 널리 사용됩니다. 아래에 Tesseract 라이브러리에 대한 주요 사항을 자세히 설명하겠습니다.

### 1. **역사 및 배경**
- **개발 초기**: Tesseract는 1985년에 Hewlett-Packard(HP) 연구소에서 개발되었습니다. 오랜 기간 동안 내부적으로 사용되다가, 2005년에 오픈 소스로 공개되었습니다.
- **Google 지원**: 2006년에 Google이 Tesseract 프로젝트를 후원하면서 활발히 발전하게 되었고, 현재 Google이 개발을 이어받아 유지하고 있습니다.

### 2. **기능 및 특징**
- **광학 문자 인식(OCR)**: Tesseract는 이미지나 스캔한 문서에서 문자를 인식하여 텍스트로 변환하는 OCR 기능을 제공합니다.
- **다양한 언어 지원**: 100개 이상의 언어를 지원하며, 한국어를 포함한 여러 언어를 인식할 수 있습니다. 사용자 정의 언어 학습을 통해 새 언어를 추가하거나 기존 언어 인식 성능을 개선할 수도 있습니다.
- **오픈 소스 라이브러리**: GNU GPL(GNU General Public License) 하에서 제공되며, 무료로 사용하고 배포할 수 있습니다.
- **이미지 전처리**: 이미지의 품질에 따라 다양한 전처리 기술이 사용될 수 있습니다. Tesseract는 이미지에서의 잡음을 줄이고 글자 인식을 향상시키기 위한 여러 이미지 전처리 기술을 지원합니다.
- **Pytesseract**: Python에서 Tesseract를 사용할 수 있는 래퍼 라이브러리입니다. 이를 통해 Python 코드 내에서 Tesseract의 기능을 쉽게 사용할 수 있습니다.

#### **Python에서 사용하기 (Pytesseract)**
Pytesseract를 사용하면 Python 코드 내에서 Tesseract의 기능을 사용할 수 있습니다.
```python
from PIL import Image
import pytesseract

# 이미지 파일 열기
image = Image.open('input_image.png')

# 이미지에서 텍스트 추출

```
- **Pillow 라이브러리**: 이미지 파일을 열고 처리하기 위해 사용됩니다.
- **`image_to_string()`**: 이미지에서 텍스트를 추출하는 기본 함수입니다.

### 3. **언어 데이터 파일**
Tesseract는 언어 인식을 위해 `.traineddata` 파일을 사용합니다. Tesseract 설치 시 기본적으로 여러 언어가 포함되어 있지만, 추가 언어가 필요한 경우 `tessdata` 폴더에 `.traineddata` 파일을 추가로 다운로드하여 넣어야 합니다.

- **언어 데이터 파일 다운로드**: [tesseract-ocr 언어 데이터 GitHub 저장소](https://github.com/tesseract-ocr/tessdata)에서 필요한 언어 데이터를 다운로드할 수 있습니다.
- **사용 예시**:
  ```bash
  tesseract input_image.png output_text -l kor
  ```
  위 명령어는 한국어를 인식하기 위해 `kor` 언어 데이터를 사용합니다.

### 4. **Tesseract의 주요 옵션**
- **`-l` 옵션**: 사용할 언어를 지정합니다. 예를 들어 `-l eng`는 영어를 의미하며, 여러 언어를 함께 사용할 수도 있습니다(`-l eng+kor`).
- **`--psm` 옵션**: 페이지 세분화 모드를 지정합니다. Tesseract는 페이지 레이아웃을 어떻게 분석할지에 대한 여러 모드를 제공합니다.
  - **예시**: `--psm 6`은 "단일 블록의 균일한 텍스트"로 간주하고 인식합니다.
- **`--oem` 옵션**: OCR 엔진 모드를 설정합니다.
  - `0`: Legacy 엔진만 사용.
  - `1`: LSTM 엔진만 사용.
  - `2`: 두 엔진 혼합 사용.
  - `3`: 기본 엔진을 자동 선택.

### 5. **Tesseract의 응용 분야**
- **문서 디지털화**: 책이나 문서를 디지털 형식으로 변환하여 데이터베이스에 저장할 때 사용됩니다.
- **명함 인식**: 명함에서 이름, 연락처 등의 정보를 추출하여 자동으로 주소록에 추가하는 데 사용됩니다.
- **자동화된 데이터 입력**: 서류나 청구서의 데이터를 자동으로 추출하여 데이터베이스에 입력하는 등 수작업을 줄이고 효율성을 높이는 데 활용됩니다.
- **차량 번호판 인식**: 차량 번호판 이미지에서 번호를 인식하여 교통 관리 시스템에 적용하는 등 실시간 분석에서 사용됩니다.

### 6. **장점과 단점**

##### 입력된 이미지를 scan한 이미지로 변경하여 ocr을 수행

````python
from imutils.perspective import four_point_transform
import numpy as np
import imutils
````

````python
# 제목(title), 이미지(img) 그리고 이미지를 표시할 때 사용 할 그림크기(figsize)를 매개변수로 받음.
def plt_imshow(title='Image', img=None, figsize=(8,5)):
    plt.figure(figsize=figsize)

    if type(img) == list:
        if type(title) == list:
            titles = title
        else:
            titles = []

            for i in range(len(img)):
                titles.append(title)

        for i in range(len(img)):
            if len(img[i].shape) <= 2:
                rgbimg = cv2.cvtColor(img[i], cv2.COLOR_GRAY2RGB)
            else:
                rgbimg = cv2.cvtColor(img[i], cv2.COLOR_BGR2RGB)

            plt.subplot(1, len(img), i+1), plt.imshow(rgbimg)
            plt.title(titles[i])
            plt.xticks([]), plt.yticks([])

        plt.show()
    else:   # 매개변수 리스트가 아닌 경우
        if len(img.shape) < 3:
            rgbimg = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
        else:
            rgbimg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.imshow(rgbimg)
        plt.title(title)
        plt.xticks([]), plt.yticks([])
        plt.show()
````

````python
url = 'https://user-images.githubusercontent.com/69428232/148330274-237d9b23-4a79-4416-8ef1-bb7b2b52edc4.jpg'

image_array = np.asarray(bytearray(requests.get(url).content), dtype=np.uint8)
org_img = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

plt_imshow("Origninal Imgae", org_img)
````

카피 한 후 변경

````python
img = org_img.copy()
img = imutils.resize(img, width=500)
ratio = org_img.shape[1] / float(img.shape[1])

# 이미지를 grayscale로 변환하고 blur를 적용
# 모서리를 찾기 위한 이미지 연산
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5,5), 0)
# 케니 가장자리 검출은 이미지에서 엣지를 감지, 이중 임계값을 사용하여 강한 엣지, 약한 엣지 및 노이즈로 분류, 75:최소 임계값, 200:최대 임계값
edged = cv2.Canny(blurred, 75, 200)

plt_imshow(["Image", "Gray", "Blurred", "Edged"], [img, gray, blurred, edged])
````

모서리 찾기

````python
# 캐니 가장자리 검출을 통해 찾은 가장자리의 윤곽을 검출.
# 함수는 이미지에서 윤곽을 찾아 리스트로 반환.
# RETR_EXTERNAL 플래그를 사용하여 가장 바깥쪽 윤곽만 검출하고, CHAIN_APPROX_SIMPLE 플래그를 사용하여 윤곽을 간단한 방법으로 압축.
cnts = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# openCV의 버전에 따라 반환되는 값의 차이를 처리. 이 함수를 사용하면 윤곽을 안전하게 추출.
cnts = imutils.grab_contours(cnts)
# 윤곽을 크기순으로 정렬.
cnts = sorted(cnts, key=cv2.contourArea, reverse=True)

# 영수증 윤곽을 저장할 변수들을 초기화.
receiptCnt = None

# 정렬된 contours를 반복문으로 수행하며 4개의 꼭지점을 갖는 도형을 검출.
for c in cnts:
    # 윤곽의 둘레를 계산
    peri = cv2.arcLength(c, True)
    # 윤곽을 단순화(근사화). 0.02 + peri는 근사화 정확도를 조절하는 파라미터.
    approx = cv2.approxPolyDP(c, 0.02 * peri, True)

    # contours가 크기순으로 정렬되어 있기 때문에 제일 첫번째 사각형을 영수증 영역으로 판단하고 break.
    if len(approx) == 4:
        receiptCnt = approx
        break

# 만약 추출한 윤곽이 없을 경우
if receiptCnt is None:
    raise Exception(("Could not find receipt outline"))

output = img.copy()
# 이미지에 윤곽을 그리는 함수. -1로 설정하면 모든 윤곽을 그림.
cv2.drawContours(output, [receiptCnt], -1, (255, 255, 0), 2)
plt_imshow("Receipt Outline", output)
````

````python
from imutils.perspective import four_point_transform
# four_point_transform() gkatnsms OpenCV를 사용한 이미지 처리에서 이미지의 특정 영역을 네 개의 꼭짓점을 기준으로 변환하는 데 사용
# 원본 이미지에 찾은 윤곽을 기준으로 이미지를 보정
# 원본 이미지에서 영수증 영역을 추출하고 변환
receipt = four_point_transform(org_img, receiptCnt.reshape(4,2) * ratio)
plt_imshow("Receipt Transform", receipt)
````

- psm 4는 Tesseract에 이미지가 "변수 크기의 텍스트를 포함하는 단일 열"로 구성되어 있다고 알려주는 모드입니다.
- Tesseract에는 여러 가지 psm 모드가 있는데, 4는 텍스트의 레이아웃이 하나의 열로 구성된 경우에 적합합니다.
  - psm 3: 자동 페이지 세그멘테이션. OCR 기본 모드입니다.
  - psm 6: 단락 텍스트를 인식하는 모드입니다.
- psm 모드를 적절히 설정하는 것은 이미지에 있는 텍스트의 구조에 맞춰 OCR의 성능을 개선할 수 있습니다.

````python
import pytesseract
import cv2
import os
from PIL import Image
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
````

````python
from imutils.perspective import four_point_transform # 이미지에서 특정 영역을 직사각형으로 변환하는 데 사용
import imutils # 이미지 처리를 단순화하고 코드를 더 간결하게 만드는 데 사용
import re
import requests
````

````python
def plt_imshow(title='Image', img=None, figsize=(8,5)):
    plt.figure(figsize=figsize)

    if type(img) == list:
        if type(title) == list:
            titles = title
        else:
            titles = []

            for i in range(len(img)):
                titles.append(title)

        for i in range(len(img)):
            if len(img[i].shape) <= 2:
                rgbimg = cv2.cvtColor(img[i], cv2.COLOR_GRAY2RGB)
            else:
                rgbimg = cv2.cvtColor(img[i], cv2.COLOR_BGR2RGB)

            plt.subplot(1, len(img), i+1), plt.imshow(rgbimg)
            plt.title(titles[i])
            plt.xticks([]), plt.yticks([])

        plt.show()
    else:   # 매개변수 리스트가 아닌 경우
        if len(img.shape) < 3:
            rgbimg = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
        else:
            rgbimg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.imshow(rgbimg)
        plt.title(title)
        plt.xticks([]), plt.yticks([])
        plt.show()
````

````python
import numpy as np

url = 'https://user-images.githubusercontent.com/69428232/155486780-55525c3c-8f5f-4313-8590-dd69d4ce4111.jpg'

img_array = np.asarray(bytearray(requests.get(url).content), dtype=np.uint8)
org_img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

plt_imshow("Original Image", org_img)
````

````python
def run_tesseract_ocr(img, width, ksize=(5,5), min_threshold=75, max_threshold=200, lang='eng'):
  img_list_title = []
  img_list = []

  img = imutils.resize(img, width=width)
  ratio = org_img.shape[1] / float(img.shape[1])

  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  blurred = cv2.GaussianBlur(gray, ksize, 0)
  edged = cv2.Canny(blurred, min_threshold, max_threshold)

  img_list_title = ['gray', 'blurred', 'edged']
  img_list = [gray, blurred, edged]

  # contours를 찾아 크기순으로 정렬
  cnts = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
  cnts = imutils.grab_contours(cnts)
  cnts = sorted(cnts, key=cv2.contourArea, reverse=True)

  BcardCnt = None

  # 정렬된 contours를 반복문으로 수행하며 4개의 꼭지점을 갖는 도형을 검출
  for c in cnts:
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * peri, True)

    # contours가 크기순으로 정렬되어 있기때문에 제일 첫번째 사각형을 영수증 영역으로 판단하고 break
    if len(approx) == 4:
      BcardCnt = approx
      break


  # 만약 추출한 윤곽이 없을 경우 오류
  if BcardCnt is None:
    raise Exception(("Could not find BusinessCard outline."))


  output = img.copy()
  cv2.drawContours(output, [BcardCnt], -1, (0, 255, 0), 2)

  img_list_title.append("BusinessCard Outline")
  img_list.append(output)

  # 원본 이미지에 찾은 윤곽을 기준으로 이미지를 보정
  Bcard = four_point_transform(org_img, BcardCnt.reshape(4, 2) * ratio)

  plt_imshow(img_list_title, img_list)
  plt_imshow("BusinessCard Transform", Bcard)

  options = "--psm 4" # --psm 4는 '단일 블록' 세그멘테이션 모드로 이미지에 있는 모든 텍스트를 하나의 블록으로 처리.

  text = pytesseract.image_to_string(cv2.cvtColor(Bcard, cv2.COLOR_BGR2RGB), lang=lang, config=options)

  # OCR결과 출력
  print("[INFO] OCR결과:")
  print("==================")
  print(text)
````

````python
run_tesseract_ocr(org_img, width=200, ksize=(5, 5), min_threshold=20, max_threshold=100, lang='kor+eng')
````
