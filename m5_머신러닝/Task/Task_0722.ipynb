```python
import pandas as pd

# 1. 데이터 로드
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
           'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
           'hours-per-week', 'native-country', 'income']
data = pd.read_csv(url, header=None, names=columns, na_values='?', skipinitialspace=True)
```

```python
# 3. 파생변수 생성
# 자본 이득 및 손실 그룹 통합
def capital_gain_loss_group(row):
    if row['capital-gain'] > 0:
        return 'Gain'
    elif row['capital-loss'] > 0:
        return 'Loss'
    else:
        return 'No_gain_or_loss'

data['capital_gain_loss_group'] = data.apply(capital_gain_loss_group, axis=1)

# 나이 그룹
bins = [0, 25, 40, 60, 100]
labels = ['Youth', 'Middle_Aged', 'Senior', 'Elderly']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels)

# 근무 시간 그룹
bins = [0, 20, 40, 60, 100]
labels = ['Part_time', 'Full_time', 'Over_time', 'Excessive']
data['hours_per_week_group'] = pd.cut(data['hours-per-week'], bins=bins, labels=labels)
```

```python
# 4. 범주형 변수 인코딩
data['sex'] = data['sex'].map({'Male': 0, 'Female': 1})
data['marital-status'] = data['marital-status'].map({
    'Married-civ-spouse': 0, 'Divorced': 1, 'Never-married': 2,
    'Separated': 3, 'Widowed': 4, 'Married-spouse-absent': 5, 'Married-AF-spouse': 6
})
# 원핫 인코딩
data = pd.get_dummies(data, columns=['race', 'age_group', 'hours_per_week_group', 'capital_gain_loss_group', 'occupation', 'relationship'])
```

```python
# 5. 이상치 제거
numerical_cols = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
for col in numerical_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
```

```python
# 6. 변수 선택 및 독립 변수 종족변수 분리
selected_features = ['sex', 'marital-status', 'education-num', ] + \
                    [col for col in data.columns if 'race_' in col] + \
                    [col for col in data.columns if 'age_group_' in col] + \
                    [col for col in data.columns if 'hours_per_week_group_' in col] + \
                    [col for col in data.columns if 'capital_gain_loss_group_' in col] + \
                    [col for col in data.columns if 'occupation' in col] + \
                    [col for col in data.columns if 'relationship' in col]

X = data[selected_features]
y = data['income'].map({'<=50K': 0, '>50K': 1})
```

```python
# 7. 데이터 표준화
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 8. 데이터 셋을 학습용과 테스트용으로 나눔
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# 9. Logistic Regression 모델 생성 및 학습
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# 10. 예측 및 평가
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# 1. 데이터 로드
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
           'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
           'hours-per-week', 'native-country', 'income']
data = pd.read_csv(url, header=None, names=columns, na_values='?', skipinitialspace=True)


# 2. 결측치 처리
data = data.dropna()

# 3. 파생변수 생성
# 자본 이득 및 손실 그룹 통합
def capital_gain_loss_group(row):
    if row['capital-gain'] > 0:
        return 'Gain'
    elif row['capital-loss'] > 0:
        return 'Loss'
    else:
        return 'No_gain_or_loss'

data['capital_gain_loss_group'] = data.apply(capital_gain_loss_group, axis=1)

# 나이 그룹
bins = [0, 25, 40, 60, 100]
labels = ['Youth', 'Middle_Aged', 'Senior', 'Elderly']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels)

# 근무 시간 그룹
bins = [0, 20, 40, 60, 100]
labels = ['Part_time', 'Full_time', 'Over_time', 'Excessive']
data['hours_per_week_group'] = pd.cut(data['hours-per-week'], bins=bins, labels=labels)


# 4. 범주형 변수 인코딩
data['sex'] = data['sex'].map({'Male': 0, 'Female': 1})
data['marital-status'] = data['marital-status'].map({
    'Married-civ-spouse': 0, 'Divorced': 1, 'Never-married': 2,
    'Separated': 3, 'Widowed': 4, 'Married-spouse-absent': 5, 'Married-AF-spouse': 6
})
# 원핫 인코딩
data = pd.get_dummies(data, columns=['race', 'age_group', 'hours_per_week_group', 'capital_gain_loss_group', 'occupation', 'relationship'])


# 5. 이상치 제거
numerical_cols = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
for col in numerical_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]


# 6. 변수 선택 및 독립 변수 종족변수 분리
selected_features = ['sex', 'marital-status', 'education-num', ] + \
                    [col for col in data.columns if 'race_' in col] + \
                    [col for col in data.columns if 'age_group_' in col] + \
                    [col for col in data.columns if 'hours_per_week_group_' in col] + \
                    [col for col in data.columns if 'capital_gain_loss_group_' in col] + \
                    [col for col in data.columns if 'occupation' in col] + \
                    [col for col in data.columns if 'relationship' in col]

X = data[selected_features]
y = data['income'].map({'<=50K': 0, '>50K': 1})

# 7. 데이터 선택 및 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# 8. 데이터 셋을 학습용과 테스트용으로 나눔
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)


# 9. Logistic Regression 모델 생성 및 학습
model = LogisticRegression()
model.fit(X_train, y_train)


# 10. 예측 및 평가
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
