## 프롬프트 엔지니어링

정의
- Prompt engineering은 인공지능 모델, 특히 언어 모델에게 특정 작업을 수행하도록 지시하는 명령어(prompt)를 설계하고 조정하는 기술입니다.
- 사용자는 명확하고 구체적인 질문이나 지시를 통해 모델이 원하는 결과를 생성하도록 유도할 수 있습니다. 이 과정은 자연어 처리에서 중요한 역할을 하며, 모델의 출력을 최적화하고 정확도를 높이는 데 기여합니다.

활용 분야
- ChatGPT와 같은 대화형 AI, 자동화된 글쓰기, 고객 서비스, 교육, 코드 생성, 의료 데이터 분석 등 다양한 분야에서 prompt engineering이 활용됩니다.
- 이를 통해 AI가 특정 요구 사항에 맞춘 결과를 생성하도록 하여, 작업 효율성을 높이고 사용자 경험을 향상시킬 수 있습니다.

사례들은 대규모 언어 모델(LLM)을 활용하는 다양한 방법을 보여줍니다. 각 사례는 특정 작업에 맞게 프롬프트를 설계하여 모델이 어떻게 유용한 결과를 생성하는지 설명합니다.

- 텍스트 생성: 창의적인 글쓰기, 마케팅 카피 작성, 기술 문서 생성 등 다양한 분야에서 사용됩니다. 프롬프트의 구체성과 맥락 설정에 따라 생성되는 텍스트의 품질이 달라집니다.

- 코드 생성 및 보완: 개발자가 원하는 코드 스니펫을 생성하거나 오류를 수정하는 데 도움을 줍니다. 예를 들어, 특정 함수나 알고리즘을 구현할 때 LLM을 활용해 코드를 제안받을 수 있습니다.

- 데이터 분석 및 요약: 대량의 데이터를 요약하거나 분석 결과를 해석하는 데 유용합니다. LLM을 통해 데이터를 자연어로 쉽게 설명할 수 있어, 비전문가도 결과를 이해하기 쉽게 만듭니다.

- 교육 및 학습 지원: 특정 주제에 대한 설명을 제공하거나, 학습 문제를 풀 때 힌트를 제시하는 방식으로 활용됩니다. 이는 교육용 도구나 학습 지원 시스템에서 유용하게 쓰일 수 있습니다.

- 대화형 응용: 고객 서비스, 개인 비서 역할을 하는 챗봇 등에서 LLM이 활용됩니다. 사용자의 질문에 대해 신속하고 정확한 답변을 제공하며, 자연스러운 대화를 통해 사용자 경험을 향상시킵니다.

- 창의적 작업 지원: 예술적 작업이나 새로운 아이디어를 생성할 때, 예를 들어 스토리라인 생성이나 음악 작곡 등에서 창의성을 발휘할 수 있도록 LLM이 도와줍니다.

## 더 나은 결과를 얻기 위한 프롬프트 6가지 전략
대규모 언어 모델(때로는 GPT 모델이라고도 함)에서 더 나은 결과를 얻기 위한 프롬프트 전략과 전술. 여기에 설명된 방법은 더 큰 효과를 위해 때때로 조합하여 사용

명확하고 구체적인 언어 사용
- 모호한 표현 대신 구체적 단어 사용: 일반적 표현보다 구체적인 단어를 사용하여 명확히 전달합니다. 예: "빠르다" 대신 "5분 내에".
- 명확한 문장 구조 유지: 단문을 사용하여 문장이 복잡하지 않도록 합니다.

단계별 지침 제공
- 작업 분할: 작업을 여러 단계로 나누어 각 단계를 명확히 설명합니다.
- 단계별 결과 명시: 각 단계에서 기대되는 결과를 구체적으로 작성합니다.

필요한 컨텍스트 제공
- 관련 배경 정보 포함: 작업을 이해하는 데 필요한 배경 정보를 제공합니다.
- 구체적 시간/상황 명시: 언제 또는 어떤 상황에서 작업이 수행되는지 명시합니다.

예시 제공
- 긍정적/부정적 예시 포함: 올바른 예시와 잘못된 예시를 모두 제공하여 비교할 수 있게 합니다.
- 다양한 시나리오 제시: 다양한 상황에 대한 예시를 포함합니다.

부가 정보 제공
- 추가 리소스 링크 제공: 참고할 수 있는 추가 자료나 링크를 제공합니다.
- 참조 자료 목록화: 관련 자료들을 목록으로 정리하여 쉽게 접근할 수 있게 합니다.

피드백 및 조정
- 피드백 요청: 지침의 명확성을 확인하기 위해 사용자로부터 피드백을 요청합니다.
- 지침 조정: 받은 피드백을 반영하여 지침을 수정하고 개선합니다.

**명확하고 구체적인 언어 사용**

모호한 표현 대신 구체적 단어 사용
- 모호한 표현: "빠르게 제출하세요."
- 구체적 표현: "5분 내에 제출하세요."

모호한 표현인 "빠르게"는 사람마다 다르게 해석될 수 있습니다. 어떤 사람에게는 "빠르게"가 1시간을 의미할 수도 있고, 다른 사람에게는 몇 분을 의미할 수도 있습니다. 반면 "5분 내에"는 정확한 시간을 명시하여 모두가 같은 의미로 이해할 수 있게 합니다. 이를 통해 지침의 명확성이 높아져 오해가 줄어들고, 작업이 더 정확하게 수행될 수 있습니다.

**명확한 문장 구조 유지**
- 복잡한 문장: "프로젝트 파일을 검토하고, 오류를 수정하며, 다음 회의 전에 팀원들과 공유하세요."
- 단순한 문장: "프로젝트 파일을 검토하세요. 오류를 수정하세요. 다음 회의 전에 팀원들과 공유하세요."

복잡한 문장은 여러 작업을 한 문장에 포함하여 읽는 사람이 이해하기 어렵게 만듭니다. 단순한 문장은 각 작업을 별도의 문장으로 분리하여 각 지시사항을 명확하게 전달합니다. 이를 통해 독자는 각 작업의 순서와 중요성을 쉽게 파악할 수 있으며, 오해나 실수를 줄일 수 있습니다.

**작업 분할**
- 복잡한 작업 지시: "보고서를 작성하고 검토한 후, 이메일로 보내세요."
- 단계별 작업 지시:<br>
  - 보고서 작성:
    - 관련 데이터를 수집하세요.
    - 수집한 데이터를 분석하여 요약하세요.
    - 분석 결과를 바탕으로 보고서를 작성하세요.
  - 보고서 검토:
    - 작성한 보고서를 읽고 오류를 수정하세요.
    - 팀원에게 보고서를 공유하고 피드백을 받으세요.

  - 이메일로 전송:
    - 검토한 보고서를 이메일에 첨부하세요.
    - 적절한 수신자에게 이메일을 보내세요.

작업을 여러 단계로 분할하면 각 단계의 구체적인 작업이 명확해집니다. 이를 통해 사용자는 각 단계를 하나씩 완료하면서 전체 작업을 체계적으로 수행할 수 있습니다. 작업 분할은 복잡한 작업을 관리 가능하고 이해하기 쉽게 만듭니다.

**단계별 결과 명시**
- 작업 지시: 보고서를 작성하고 검토한 후, 이메일로 보내세요.
- 단계별 결과 명시:
  - 보고서 작성:
    - 결과: 초안 보고서 완성. 데이터가 정확히 포함되어 있음.
  - 보고서 검토:
    - 결과: 오류가 수정된 보고서. 피드백이 반영됨.
  - 이메일로 전송:
    - 결과: 보고서가 이메일에 첨부되어 적절한 수신자에게 전송됨.

각 단계에서 기대되는 결과를 구체적으로 명시하면 사용자가 작업의 목표를 명확히 이해할 수 있습니다. 이를 통해 단계별로 작업의 완성도를 확인할 수 있고, 최종 결과물이 기대에 부합하는지 확인할 수 있습니다. 단계별 결과 명시는 작업의 명확성과 효율성을 높이는 데 중요한 역할을 합니다.

**관련 배경 정보 포함**
- 지시사항: "지난 주의 프로젝트 파일을 검토하세요."
- 관련 배경 정보 포함:
  - 프로젝트 개요:
    - 배경 정보: "이 프로젝트는 고객 만족도 향상을 목표로 합니다. 지난 주 회의에서 고객 피드백을 분석하고, 이를 반영한 새로운 전략을 논의했습니다."
  - 특정 파일 지시:
    - 배경 정보: "지난 주에 작성된 '고객 피드백 분석 보고서'와 '전략 수정 제안서' 파일을 검토하세요. 이 파일들은 프로젝트 폴더의 '2024년 8월' 섹션에 있습니다."
  - 검토 목적:
    - 배경 정보: "이번 검토는 제안된 전략이 효과적인지 평가하고, 추가 개선 사항을 도출하는 것을 목표로 합니다."

작업을 이해하는 데 필요한 배경 정보를 제공하면 사용자가 작업의 목적과 맥락을 명확히 이해할 수 있습니다. 이로 인해 작업 수행의 정확성과 효율성이 높아집니다. 배경 정보는 작업의 중요성, 관련 문서 위치, 수행해야 할 구체적 내용 등을 포함할 수 있습니다.

**구체적 시간/상황 명시**
- 모호한 지시사항: "회의 전에 보고서를 작성하세요."
- 구체적 지시사항:
  - 시간 명시:
    - 구체적 표현: "월요일 오전 9시 회의 전에 보고서를 작성하세요."
    - 상황: "회의는 매주 월요일 오전 9시에 시작됩니다. 이 회의에서 이번 주 목표와 작업 분담이 논의될 예정입니다."
  - 상황 명시:
    - 구체적 표현: "월간 실적 검토 회의 전에, 이번 달 매출 보고서를 작성하세요."
    - 상황: "월간 실적 검토 회의는 매월 마지막 금요일 오후 3시에 열립니다. 이 보고서는 전체 팀의 실적을 평가하는 데 사용됩니다."

구체적인 시간과 상황을 명시하면 작업의 시기와 맥락을 명확히 할 수 있습니다. 이는 사용자가 언제 작업을 완료해야 하는지 정확히 이해하게 하여, 일정 준수와 작업의 적시 수행을 도와줍니다. 또한, 작업의 중요성과 맥락을 이해함으로써, 더 나은 준비와 수행이 가능해집니다.

**긍정적/부정적 예시 포함**
- 지시사항: "이메일을 작성하세요."
- 긍정적 예시:
  - 제목: "회의 일정 확인 요청"
  - 내용: "안녕하세요, 김 팀장님. 이번 주 금요일 오후 2시에 예정된 회의 일정을 확인하고자 합니다. 참석 가능하신지 알려주시면 감사하겠습니다. 감사합니다. 홍길동 드림."
- 부정적 예시:
  - 제목: "회의"
  - 내용: "안녕하세요, 회의가 이번 주 금요일에 있을 것 같습니다. 참석할 수 있는지 알려주세요. 홍길동."

긍정적 예시와 부정적 예시를 제공하면 사용자는 무엇이 올바르고 무엇이 잘못된지 명확히 이해할 수 있습니다. 긍정적 예시는 최상의 결과를 보여주어 기대하는 바를 명확히 전달하며, 부정적 예시는 피해야 할 실수를 구체적으로 보여줍니다. 이를 통해 사용자는 지침을 더 잘 따르고, 결과물의 품질을 높일 수 있습니다.

**다양한 시나리오 제시**
- 지시사항: "보고서를 작성하세요."
- 다양한 시나리오:
  - 주간 보고서 작성:
    - 상황: 매주 금요일마다 주간 업무 요약 보고서를 작성합니다.
    - 포함 내용: 주요 성과, 문제점, 다음 주 계획.
    - 긍정적 예시: "이번 주 주요 성과로는 판매 목표 120% 달성이 있습니다. - 문제점으로는 재고 부족이 있었으며, 이를 해결하기 위해 다음 주에는 재고 보충 계획을 세웠습니다."
    - 부정적 예시: "이번 주에 여러 가지 일이 있었고, 다음 주에도 할 일이 많습니다."
  - 프로젝트 보고서 작성:
    - 상황: 프로젝트 마감 시 프로젝트 진행 상황과 결과를 요약한 보고서를 작성합니다.
    - 포함 내용: 프로젝트 목표, 진행 상황, 최종 결과, 배운 점.
    - 긍정적 예시: "프로젝트 목표는 고객 만족도 향상이었고, 진행 상황은 계획대로 진행되었습니다. 최종 결과로 고객 만족도가 20% 향상되었습니다. 배운 점은 초기 계획의 중요성입니다."
    - 부정적 예시: "프로젝트가 끝났고, 결과가 나왔습니다."
  - 연구 보고서 작성:
    - 상황: 연구 프로젝트의 결과를 발표하기 위한 보고서를 작성합니다.
    - 포함 내용: 연구 목적, 방법, 결과, 결론.
    - 긍정적 예시: "연구 목적은 새로운 항생제의 효과를 검증하는 것이었습니다. 실험 방법은 무작위 대조 실험이었고, 결과는 긍정적이었습니다. 결론은 새로운 항생제가 효과적이라는 것입니다."
    - 부정적 예시: "연구를 했고, 결과가 좋았습니다."

다양한 시나리오를 제시하면 사용자는 각 상황에 맞게 작업을 어떻게 수행해야 하는지 명확히 이해할 수 있습니다. 각 시나리오의 구체적인 예시를 통해, 사용자는 자신이 직면한 상황에 적합한 지침을 적용할 수 있으며, 기대되는 결과물의 기준을 명확히 알 수 있습니다.

**추가 리소스 링크 제공 및 참조 자료 목록화**
- 지시사항: "보고서를 작성하기 위한 참고 자료를 확인하세요."
- 추가 리소스 링크 제공:
  - 링크 1: 보고서 작성 가이드라인
  - 링크 2: 데이터 분석 기법
  - 링크 3: 피드백 수집 방법
- 참조 자료 목록화:

피드백 요청을 통해 지침의 명확성을 확인하고 개선할 수 있습니다. 사용자로부터 피드백을 받으면, 지침이 제대로 이해되고 있는지 확인할 수 있으며, 필요한 경우 수정하여 더 명확하고 효과적인 지침을 제공할 수 있습니다. 이를 통해 작업의 품질과 정확성을 높일 수 있습니다.

반복과 테스트
- 여러 프롬프트를 테스트하여 최적의 결과 도출

```python
prompts = [
    "정기적인 운동의 건강상의 이점은 무엇인가요?",
    "정기적으로 운동을 하는 것의 장점을 설명해주세요",
    "정기적인 신체 활동이 건강에 어떤 영향을 미치나요?"
]

for prompt in prompts:
    completion = client.chat.completions.create(
        model="gpt-4o-mini-2024-07-18",
        messages=[
            {"role": "system", "content": "당신은 건강관리 전문가이다."},
            {"role": "user", "content": prompt}
    ]
)
    print(f"프롬프트: {prompt}")
    print(completion.choices[0].message.content)
    print("\n" + "="*100 + "\n")
```
