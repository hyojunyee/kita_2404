```python
import numpy as np
import gradio as gr
from skimage.transform import resize

def sepia(input_img):
    input_img = resize(input_img, (224, 224))
    sepia_filter = np.array([
        [0.393, 0.769, 0.189],
        [0.349, 0.686, 0.168],
        [0.272, 0.534, 0.131]
    ])
    sepia_img = input_img.dot(sepia_filter.T)
    sepia_img /= sepia_img.max()
    return sepia_img

# ì¸í„°í˜ì´ìŠ¤ ì •ì˜ ì‹œ shape ë§¤ê°œë³€ìˆ˜ ì œê±°
demo = gr.Interface(
    fn=sepia,
    inputs=gr.Image(),
    outputs="image")

demo.launch()
```

```python
import gradio as gr

def bmi(name, height, weight, feeling):
    bmi_val = round(weight / ((height / 100) **2), 2)
    result_emoji = "ğŸ¤£" if bmi_val < 30 else "ğŸ¤”"
    output_str = "Hello" + name + "... \n your BMI is ..." + str(bmi_val)
    text = "Happy" if feeling else "Sad"
    return (output_str, result_emoji, text)

bmi(" hyojun", 172, 70, True)
```
