Q. 주어진 문자열 리스트에서 유효한 Python 변수 이름만을 추출하여 반환하는 함수를 작성하세요.

```python
identifiers = ["var1", "2things", "variabel_name", "time!", "True"]
```

```python
#1. 제외시켜가며 하는 법.
#제거되는 시점에서 인덱스가 변경됨(인덱스는 리스트 순서)
import keyword
def extract_valid_identifiers(identifiers):
    valid_identifiers = []
    for identifier in identifiers:
        if identifier[0].isdigit():
            continue
        is_valid = all(char.isalnum() or char == '_' for char in identifier)
        if is_valid and identifier not in keyword.kwlist:
            valid_identifiers.append(identifier)
    return valid_identifiers

identifiers = ["var1", "2things", "variabel_name", "time!", "True"]
valid_identifiers = extract_valid_identifiers(identifiers)
print(valid_identifiers)
```

```python
#2. 골라가며 하는법
#isidentifier()는 파이썬의 문자열 메서드로, 주어진 문자열이 유요한 식별자인지를 확인
#isidentifier()는 예약어인지는 따로 체크하지 않음
import keyword
def valid_identifiers(idnetifiers):
    valid = []
    for identifier in identifiers:
      if identifier.isidentifier() and not keyword.iskeyword(identifier):
         valid.append(identifier)
    return valid

#예제 실행
identifiers = ["var1", "2things", "variabel_name", "time!", "True"]
print(valid_identifiers(identifiers))
```

#### sort 와 sorted 차이점

1. 메서드 vs 함수:
- sort(): 리스트 객체의 내장 메서드입니다. 즉, 리스트에서만 사용할 수 있습니다.

예를 들어, ['a', 'a', 'b', 'c', 'c', 'c', 'd', 'e', 'e']가 입력되면, ['a', 'b', 'c', 'd', 'e']를 출력해야 합니다.

```python
list1 =  ['a', 'a', 'b', 'c', 'c', 'c', 'd', 'e', 'e']
result = [list1[0]]  # ['a'] -> 초기값. 반복문을 통해 추가한다.
for i in list1[1:]:
  if i != result[-1]: # != -> (같지않다) 조건이 맞지않으면 실행안됨.
    result.append(i)
print(result)
```

```python
def remove_consecutive_duplicates(items):
# items 가 없으면 결과값이 없어짐 -> 리스트가 없을때 에러가 뜨는걸 방지
  # if not items:
  #   return[]
  # 결과리스트 초기화
  result = [items[0]]
  # 이전 요소와 비교하면서 중복 제거
  for cur in items[1:]:
    if cur != result[-1]:
      result.append(cur)
  return result

# 리스트 정의 후 출력
list1 = ['a', 'a', 'b', 'c', 'c', 'c', 'd', 'e', 'e']
print(remove_consecutive_duplicates(list1))
```

Task5_0425 주어진 정수 리스트와 회전 횟수 k에 대해 리스트를 오른쪽으로 k만큼 회전시킨 결과를 반환하세요. k가 리스트의 길이보다 클 수 있으며, 이 경우 k를 리스트 길이로 나눈 나머지만큼 실제 회전시키면 됩니다.

예를 들어, [1, 2, 3, 4, 5]와 k=2가 주어지면, 결과는 [4, 5, 1, 2, 3]이 되어야 합니다.

```python
def rotate_list(list1, k):
  # if not list1:
  #   return list1
  # k가 리스트 길이보다 클 수 있기때문
  k = k % len(list1)
  # 회전 횟수를 리스트 길이로 나눈 나머지로 계산
  rotated = list1[-k:] + list1[:-k]
  # 리스트의 끝부분을 앞으로, 앞부분을 끝으로 옮김
  return rotated

# 리스트 정의
list1 = [1, 2, 3, 4, 5]
k = 2
# 결과값 출력
print(rotate_list(list1, k))
```

```python
def rerotate_list(list1, k):
  # if not list1:
  #   return list1
  # k가 리스트 길이보다 클 수 있기때문
  k = k % len(list1)
  # 리스트의 끝부분을 앞으로, 앞부분을 끝으로 옮김.
  rerotated = list1[k:] + list1[:k]
  return rerotated

# 리스트 정의
list1 = [1, 2, 3, 4, 5]
k = 2
# 결과값 출력
print(rerotate_list(list1, k))
```

```python
class AddMul: #self를 넣어줘야된다.
  def add(self, n1, n2):
    return n1 + n2
  def mul(self, n1, n2):
    return n1 * n2

obj1 = AddMul()
print(obj1.add(1,1))
obj1 = AddMul()
print(obj1.mul(3,4))
```

Task6_0426. Calculator 클래스를 작성하고 4칙연산을 수행하는 객체 4개를 작성하여 결과를 출력하세요.

```python
class Calculator:
# 두 파라메타값의 덧셈
  def add(self, n1, n2):
    return n1 + n2
# 두 파라메타값의 뺄셈
  def subt(self, n1, n2):
    return n1 - n2
# 두 파라메타값의 곱셈
  def mul(self, n1, n2):
    return n1 * n2
# 두 파라메타값의 나눗셈
  def div(self, n1, n2):
    if n1 == 0 or n2 == 0:
      return "Error : Division by zero"
    else:
      return round(n1 / n2)

obj1 = Calculator()
print(obj1.add(1,2))
obj2 = Calculator()
print(obj2.subt(3,4))
obj3 = Calculator()
print(obj3.mul(5,9))
obj4 = Calculator()
print(obj1.div(9,3))
obj4 = Calculator()
print(obj1.div(2,0))
```
