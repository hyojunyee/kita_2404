```python
from pymongo import MongoClient
from openai import OpenAI

# MongoDB 서버에 연결
client_db = MongoClient('mongodb://localhost:27017/')
db = client_db['team3']
collection = db['team3']

# OpenAI API 클라이언트 초기화
client_ai = OpenAI(api_key="")

# 시스템 메시지 (OpenAI의 초기 설정)
system = """
    당신은 노동법에 관한 다양한 질문에 답변할 수 있는 전문가 챗봇입니다.
    - 사용자의 질문에 대해 정확하고 상세한 답변을 제공해야 합니다.
    - 법률에 관한 db는 사용자가 법률 조항을 요청할 때 참고할 수 있도록 사용합니다.
    - 답변이 부정확할 가능성이 있을 경우, 해당 내용을 잘 모른다고 답변합니다.
    - 질문이 구체적이지 않은 경우, 추가적인 정보를 요청하여 명확한 답변을 제공하는 것이 좋습니다.
    - 각 질문에 대해 관련 법률 조항이 필요할 때는 데이터베이스를 검색하여 제공합니다.
"""

# 대화 히스토리 초기화
messages = [
    {"role": "system", "content": system}
]

def find_law_clause(keyword):
    """
    주어진 키워드를 사용하여 법률 조항을 검색하고, 그 내용을 반환합니다.
    """
    query = {"content": {"$regex": keyword, "$options": "i"}}  # 대소문자 구분 없이 검색
    results = collection.find(query)
    
    clauses = []
    for result in results:
        clauses.append(result["content"])
    
    return clauses

def generate_response(question, clauses):
    """
    OpenAI API를 호출하여 사용자의 질문에 대한 응답을 생성합니다.
    """
    # 사용자 질문에 법률 조항 추가
    if clauses:
        clause_content = "\n".join(clauses)
        messages.append({"role": "user", "content": f"관련 법조항: {clause_content}"})
    else:
        messages.append({"role": "user", "content": f"질문: {question}"})
    
    # OpenAI API 호출
    response = client_ai.chat.completions.create(
        model="gpt-4o-mini-2024-07-18",
        messages=messages,
        temperature=0.3,
        top_p=0.2
    )
    
    # AI의 응답을 대화 히스토리에 추가
    reply = response.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})
    
    return reply

def request_more_info(question):
    """
    질문에 대한 추가 정보를 요청합니다.
    """
    additional_questions = [
        "질문하신 주제와 관련하여 좀 더 구체적인 정보를 제공해주실 수 있나요? 예를 들어, 특정 날짜나 급여 조건 등",
        "해당 상황과 관련하여 추가적인 세부사항이 필요한가요?",
        "이와 관련된 법률 조항이나 규정에 대해 구체적으로 알고 싶으신 부분이 있나요?"
    ]
    
    messages.append({"role": "user", "content": f"기타 추가 질문: {', '.join(additional_questions)}"})
    
    # OpenAI API 호출
    response = client_ai.chat.completions.create(
        model="gpt-4o-mini-2024-07-18",
        messages=messages,
        temperature=0.3,
        top_p=0.2
    )
    
    # AI의 응답을 대화 히스토리에 추가
    reply = response.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})
    
    return reply

while True:
    # 사용자의 질문 입력 받기
    question = input("질문을 입력해주세요 (종료하려면 'exit' 입력): ")
    
    # 종료 조건
    if question.lower() == "exit":
        break
    
    # 질문이 구체적이지 않은 경우 추가 정보 요청
    if any(word in question for word in ["휴가", "급여", "근로", "수당", "복지"]):
        reply = request_more_info(question)
    else:
        # MongoDB에서 법률 조항 검색
        keyword = question.split()[0]  # 예: "휴가"를 키워드로 추출
        law_clauses = find_law_clause(keyword)
        
        # OpenAI API를 통해 응답 생성
        reply = generate_response(question, law_clauses)
    
    # AI의 응답 출력
    print(reply)
```
